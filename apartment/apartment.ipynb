# Нормировка данных

Семён окончил Университет и решил снять себе квартиру. Семён был готов проехать несолько остановок на метро. Конечо, лучше, если *этих остановок* будет поменьше. Но от метро до квартиры Семёну было принципиально добираться пешком -- не более 20 минут. И, конечно, хотелось снять квартиру поближе к метро.

Состояние квартиры Семёна интересовало не слишком, но он готов был посмотреть 3 возможных варианта. Последний важный момент -- цена вопроса. Все факторы показались Семёну равнозначными.

**Вариант (15 mod 6) + 1 = 4**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("apartment-dataset-4.csv", delimiter=";")
df.head(10)
```

ID -- ID квартиры

DISTANCE -- время, которое нужно потратить, чтобы дойти от дома до метро

STOP_COUNT -- количество остановок на метро

COST -- стоимость аренды в месяц (в тысячах рублей)

```python
# Нормализация с помощью линейной нормировки: (x - xmin) / (xmax - xmin)

for col in ['DISTANCE', 'STOP_COUNT', 'COST']:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
df['RATE'] = df['DISTANCE'] + df['STOP_COUNT'] + df['COST']
df
```

```python
# Выбор трёх наилучших вариантов

df.sort_values('RATE', ascending=False).head(3)
```
