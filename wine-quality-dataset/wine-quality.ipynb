# Wine Quality Data Set

**Abstract**: Two datasets are included, elated to red and white vinho verde wine samples, from the north of Portugal.


## Attribute Information

**Attributes based on physicochemical tests:**
1. fixed acidity
2. volatile acidity
3. citric acid
4. residual sugar
5. chlorides
6. free sulfur dioxide
7. total sulfur dioxide
8. density
9. pH
10. sulphates
11. alcohol

**Attributes based on sensory data:**
12. quality (score between 0 and 10)

```python
import pandas as pd
df = pd.read_csv("winequality-white.csv", delimiter=";")
df[10:20]
```

```python
chlorides = df['chlorides']

print("Статистика:")
print("\tВсего экземпляров: %d" % chlorides.describe()['count'])
print("\tСреднее содержание хлоридов: %.2f" % chlorides.describe()['mean'])
print("\tМинимальное содержание хлоридов: %.2f" % chlorides.describe()['min'])
print("\tМаксимальное содержание хлоридов: %.2f" % chlorides.describe()['max'])
print("\tСтандартное отклонение: %.2f" % chlorides.describe()['std'])
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt; plt.rcParams.update({'figure.figsize':(20,8), 'figure.dpi':200})

plt.hist(chlorides, bins=340)
plt.gca().set(ylabel='Frequency', xlabel="Chlorides content")
plt.show()
```

```python
std = chlorides.describe()['std']
mean = chlorides.describe()['mean']
newset = df[abs(df['chlorides'] - mean) <= 3*std]['chlorides']
plt.hist(newset, bins=102)
plt.show()
```

```python
print("Статистика:")
print("\tВсего экземпляров: %d" % newset.describe()['count'])
print("\tСреднее содержание хлоридов: %.2f" % newset.describe()['mean'])
print("\tМинимальное содержание хлоридов: %.2f" % newset.describe()['min'])
print("\tМаксимальное содержание хлоридов: %.2f" % newset.describe()['max'])
print("\tСтандартное отклонение: %.2f" % newset.describe()['std'])
```
