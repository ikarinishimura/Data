```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

df_male = pd.read_csv('male.csv', sep=';', encoding = "cp1251")
df_female = pd.read_csv('female.csv', sep=';', encoding = "cp1251")
df = pd.concat([df_male, df_female])
df = df[['Name', 'NumberOfPersons', 'Month', 'Year']]
df
```

```python
data = df[df.Name == 'Артём']
data.head(5)


sns.boxplot(x='Name', y='NumberOfPersons', data=data)
fig = plt.gcf()
fig.set_size_inches(8, 6)
```

```python
data = df[df.Name == 'Андрей']
data.head(5)

import seaborn as sns
sns.boxplot(x='Name', y='NumberOfPersons', data=data)
fig = plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

```python
data = df[df.Name == 'Елена']
data.head(5)

import seaborn as sns
sns.boxplot(x='Name', y='NumberOfPersons', data=data)
fig = plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

```python
data.describe()
```

```python
Q3 = data.describe().NumberOfPersons['75%']
Q1 = data.describe().NumberOfPersons['25%']
IQR = Q3 - Q1
print("Нижний квартиль: %.2f" % Q3)
print("Нижний квартиль: %.2f" % Q1)
print("Размах: %.2f" % IQR)
print("")
print("Умеренные выбросы:")
data[(data.NumberOfPersons - Q3 > 1.5 * IQR) & (data.NumberOfPersons <= Q3 + 3 * IQR)]
```

```python
print("Умеренные выбросы:")
data[data.NumberOfPersons - Q3 > 3 * IQR]
```
